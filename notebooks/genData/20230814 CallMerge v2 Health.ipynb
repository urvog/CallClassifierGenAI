```python
import pandas as pd
import numpy as np
import os
```

```python
#importing data

#list of all the files from ../../data
files = os.listdir("../../data/raw3_health/")

#iterating through the files and appending them to a dataframe
df = pd.DataFrame()
for file in files:
    if file.endswith(".csv"):
        #concatenating the dataframes
        df = pd.concat([df, pd.read_csv("../../data/raw3_health/"+file)])

```

```python
df
```

```python

#generating a list of 240 rows of random dates for year 2022
dates = pd.date_range(start='2021-01-01', end='2021-12-31', periods=df.shape[0])

#adding a column of random dates to the dataframe
df['Date'] = dates
df
```

```python
#saving data
df.to_csv("../../data/transcripts_health_2021.csv", index=False)
```
